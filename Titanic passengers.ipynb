```python
df["Age"].hist(bins=20)
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
survived_counts = df["Survived"].value_counts()
plt.bar(survived_counts.index, survived_counts.values)
plt.xticks([0,1], ["Not survived", "Survived"])
plt.ylabel("Count")
plt.show()
```
